````python
labels = ['country of origin', 'shareholder of', 'licensed to broadcast to', 'father', 'transferred money to', 'followed by', 'characters', 'invests in']
````

````python
from glirel.modules import custom_tokenizers

splitter = custom_tokenizers.WhitespaceTokenSplitter()

graph = json.loads(out)


relation_statements_tokens = []
ner = []
for edge in graph['edges']:
    relation_statement = f"{edge['from']} {edge['label']} {edge['to']}"
    head_tokens = list(splitter(edge['from']))
    tail_tokens = list(splitter(edge['to']))
    relation_statement_tokens = [e[0] for e in list(splitter(relation_statement))]
    relation_statements_tokens.append(relation_statement_tokens)
    ner.append(
        [
            [0, len(head_tokens)-1, 'TYPE', edge['from']],
            [len(relation_statement_tokens) - len(tail_tokens), len(relation_statement_tokens) - 1, 'TYPE', edge['to']]
        ]
    )
    
for tokens, ents in zip(relation_statements_tokens, ner):
    print(tokens)
    print(ents)
````
